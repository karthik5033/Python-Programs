```python
import pandas as pd

# Load metadata CSV
meta = pd.read_csv("HAM10000_metadata.csv")

meta.head()
```

```python
def age_group(age):
    if pd.isna(age): return "unknown"
    age = int(age)
    if age < 16: return "child"
    elif age < 60: return "adult"
    else: return "elderly"

meta['age_group'] = meta['age'].apply(age_group)

label_map = {
    'akiec': 'Actinic Keratoses',
    'bcc': 'Basal Cell Carcinoma',
    'bkl': 'Benign Keratosis',
    'df': 'Dermatofibroma',
    'mel': 'Melanoma',
    'nv': 'Nevus (Mole)',
    'vasc': 'Vascular Lesion'
}

meta['disease_name'] = meta['dx'].map(label_map)
```

```python
import os
from tqdm import tqdm

bcc = meta[meta['dx'] == 'bcc']
bcc.to_csv("bcc_list.csv", index=False)

save_folder = "datasets/bcc"
os.makedirs(save_folder, exist_ok=True)

dataset_name = "kmader/skin-cancer-mnist-ham10000"

for img_id in tqdm(bcc["image_id"]):
    os.system(f'kaggle datasets download -d {dataset_name} -f HAM10000_images_part_1/{img_id}.jpg -p "{save_folder}"')
    os.system(f'kaggle datasets download -d {dataset_name} -f HAM10000_images_part_2/{img_id}.jpg -p "{save_folder}"')
```

```python
import shutil

source_folder = "datasets/bcc"
output_zip = "bcc_dataset"   # final result will be bcc_dataset.zip

shutil.make_archive(output_zip, 'zip', source_folder)

print("BCC dataset zipped successfully!")
```

```python
import zipfile

for file in os.listdir(save_folder):
    if file.endswith(".zip"):
        zip_path = os.path.join(save_folder, file)
        try:
            with zipfile.ZipFile(zip_path, 'r') as zip_ref:
                zip_ref.extractall(save_folder)
        except:
            pass
        os.remove(zip_path)
```
