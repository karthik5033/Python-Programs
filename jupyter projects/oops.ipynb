```python
# Instance variable and methods
class Dog:
    # Constructor
    def __init__(self,name,age):
        self.name=name
        self.age=age

dog1=Dog("mf",12)
print(dog1.name)
print(dog1.age)


```

```python
class BankAccount:
    def __init__(self,owner,balance=0):
        self.owner=owner
        self.balance=balance
    def deposit(self,amount):
        self.amount=amount
        self.balance+=self.amount
        print(f"the balance of {self.owner} after adding {self.amount}rs is {self.balance}")
        
c1=BankAccount("mf",10)
c1.deposit(120)

c2=BankAccount("bf")
c2.deposit(120)
```
